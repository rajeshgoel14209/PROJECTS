```
## Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.
## Using the same function, find the team which has maximum and minimum goal difference. (5 points) 
## Goal_diff_count = GoalsFor - GoalsAgainst
```

```
def Goal_diff_count(data):
  data['GoalsFor'] = data['GoalsFor'].astype(np.int64)
  data['GoalsAgainst'] = data['GoalsAgainst'].astype(np.int64)
  GoalsFor = np.array(data['GoalsFor'])
  GoalsAgainst = np.array(data['GoalsAgainst'])
  Goal_diff =  GoalsFor - GoalsAgainst
  data['Goal_diff'] = pd.DataFrame(Goal_diff)
  return data

```

```
Goal_diff_count(df_na).max().to_frame()  ## maximum goal difference
```

```
Goal_diff_count(df_na).max().to_frame()  ## minimum goal difference
```
